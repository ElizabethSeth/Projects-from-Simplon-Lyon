### Transforming json file to csv file and keep only interesting values for future dataset

```python
with open('data.json') as f:
    data = json.load(f)
print(data)

print(type(data))
print(data.keys())
```

```python
teams_data = data['teams'] 
teams_df = pd.json_normalize(teams_data)
teams_df
```

```python
group_a_matches = data['groups']['a']['matches']
group_a_matches_df = pd.json_normalize(group_a_matches)
group_a_matches_df
```

```python
round_16_matches = data['knockout']['round_16']['matches']
round_16_matches_df = pd.json_normalize(round_16_matches)
round_16_matches_df
```

```python
team_map = {team['id']: team['name'] for team in data['teams']}
stadium_map = {stadium['id']: stadium['city'] for stadium in data['stadiums']}
matches_list = []

def process_matches(match_data, edition_name, round_name):
    for match in match_data:
        match_info = {
            'id_match': match['name'],
            'home_team': team_map[match['home_team']],
            'away_team': team_map[match['away_team']],
            'home_result': match['home_result'],
            'away_result': match['away_result'],
            'date': match['date'],
            'round': round_name,
            'city': stadium_map[match['stadium']],
            'edition': edition_name
        }
        matches_list.append(match_info)

for group_key, group_data in data['groups'].items():
    process_matches(group_data['matches'], '2018-RUSSIA', 'Group')

knockout_rounds = {
    'round_16': 'Round of 16',
    'round_8': 'Quarter-finals',
    'round_4': 'Semi-finals',
    'round_2_loser': 'Third place play-off',
    'round_2': 'Final'
}

for round_key, round_name in knockout_rounds.items():
    process_matches(data['knockout'][round_key]['matches'], '2018-RUSSIA', round_name)

data_2018 = pd.DataFrame(matches_list)

data_2018
```

### Csv File Matches from 1930-2014

```python
matches_df = pd.read_csv('matches.csv')
matches_df
```

```python
matches_df.columns
```

```python
matches_df['result'] = matches_df['result'].astype(str)

print(matches_df['result'].unique()) 
```

```python
def clean_result(result):
    match = re.search(r'(\d+-\d+)', result)
    if match:
        return match.group(1)
    else:
        return None  

matches_df['cleaned_result'] = matches_df['result'].apply(clean_result)

print(matches_df[['result', 'cleaned_result']].head())
```

```python
matches_df = matches_df.rename(columns={
    'team1': 'home_team',
    'team2': 'away_team',
    'score': 'result',
    'year': 'date',
    'round': 'round',
    'venue': 'city',
    'edition': 'edition'
})
def split_result(result):
    match = re.search(r'(\d+-\d+)', result)
    if match:
        return match.group(1).split('-')
    else:
        return [None, None]

matches_df[['home_result', 'away_result']] = matches_df['result'].apply(split_result).apply(pd.Series)

matches_df['home_result'] = pd.to_numeric(matches_df['home_result'], errors='coerce')
matches_df['away_result'] = pd.to_numeric(matches_df['away_result'], errors='coerce')
matches_df = matches_df.dropna(subset=['home_result', 'away_result'])
print(matches_df[['home_team', 'away_team', 'home_result', 'away_result' ]].head())
```
