```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.gov.uk/browse/citizenship'
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    print(soup.prettify())  
else:
    print("Failed to retrieve the page")
```

```python
cards = soup.find_all('div', class_='gem-c-cards__list-item-wrapper')

for card in cards:
    title = card.find('h2').text.strip()
    description = card.find('p').text.strip()
    link = card.find('a')['href']
    print(f"Title: {title}\nDescription: {description}\nLink: https://www.gov.uk{link}\n")
```

```python
urls = [
    {'url': 'https://www.gov.uk/browse/citizenship', 'group': 'Citizenship'},
    {'url': 'https://www.gov.uk/browse/childcare-parenting', 'group': 'Childcare and Parenting'},
    {'url': 'https://www.gov.uk/browse/births-deaths-marriages', 'group': 'Births, Deaths, Marriages'},
    {'url': 'https://www.gov.uk/browse/business', 'group': 'Business'},
    {'url': 'https://www.gov.uk/browse/justice', 'group': 'Justice'},
    {'url': 'https://www.gov.uk/browse/disabilities', 'group': 'Disabilities'},
    {'url': 'https://www.gov.uk/browse/driving', 'group': 'Driving'},
    {'url': 'https://www.gov.uk/browse/education', 'group': 'Education'},
    {'url': 'https://www.gov.uk/browse/employing-people', 'group': 'Employing-people'},
    {'url': 'https://www.gov.uk/browse/environment-countryside', 'group': 'Environment-countryside'},
    {'url': 'https://www.gov.uk/browse/housing-local-services', 'group': 'Housing-local-services'},
    {'url': 'https://www.gov.uk/browse/tax', 'group': 'Tax'},
    {'url': 'https://www.gov.uk/browse/abroad', 'group': 'Abroad'},
    
    
]
all_data = []

for entry in urls:
    response = requests.get(entry['url'])
    soup = BeautifulSoup(response.content, 'html.parser')
    
    if entry['group'] == 'Business':
        cards = soup.find_all('div', class_='gem-c-cards__list-item-wrapper')
        for card in cards:
            title = card.find('h3')
            description = card.find('p')
            link = card.find('a')
            
            if title and description and link:
                all_data.append({
                    'title': title.text.strip(),
                    'group': entry['group'],
                    'description': description.text.strip(),
                    'link': f"https://www.gov.uk{link['href']}"
                })
    else:
        cards = soup.find_all('div', class_='gem-c-cards__list-item-wrapper')
        for card in cards:
            title = card.find('h2')
            description = card.find('p')
            link = card.find('a')

            if title and description and link:
                all_data.append({
                    'title': title.text.strip(),
                    'group': entry['group'],
                    'description': description.text.strip(),
                    'link': f"https://www.gov.uk{link['href']}"
                })

df = pd.DataFrame(all_data)

print(df)

# df.to_csv('govuk_data.csv', index=False)
```

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2022_per_game.html'
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

table = soup.find('table', {'id': 'per_game_stats'})

headers = [th.text for th in table.find('thead').find_all('th')][1:]  

data = []
for row in table.find('tbody').find_all('tr', class_=lambda x: x != 'thead'):  
    cols = row.find_all('td')
    if cols:  
        player_stats = [col.text for col in cols]
        data.append(player_stats)

df = pd.DataFrame(data, columns=headers)

print(df)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

driver = webdriver.Chrome()
url = 'https://www.adamchoi.co.uk/overs/detailed'
driver.get(url)


all_matches_button = driver.find_element(By.XPATH, '//label[@analytics-event="All matches"]')
all_matches_button.click()

time.sleep(3)

matches = driver.find_elements(By.TAG_NAME, 'tr')

data = []
for match in matches:
    details = match.find_elements(By.TAG_NAME, 'td')  
    if len(details) >= 4:
        date = details[0].text
        home_team = details[1].text
        score = details[2].text
        away_team = details[3].text
        data.append({
            'Date': date,
            'Home Team': home_team,
            'Score': score,
            'Away Team': away_team
        })

df = pd.DataFrame(data)

print(df)

# df.to_csv('football_matches.csv', index=False)

driver.quit()
```
