```python
def print_first_5_documents(collection, collection_name):
    documents = collection.find().limit(5)  
    print(f"First 5 docs un this collection  {collection_name}:")
    for document in documents:
        print(document)
    print("\n")

print_first_5_documents(clients_collection, "clients")
print_first_5_documents(produits_collection, "produits")
print_first_5_documents(ventes_collection, "ventes")
```

```python
from collections import Counter

client_purchases = [doc['client_id'] for doc in ventes_collection.find()]
client_purchase_count = Counter(client_purchases)

print("Purchase frequency by client:")
for client_id, count in client_purchase_count.items():
    print(f"Client {client_id}: {count} purchases")
```

```python
product_sales = ventes_collection.aggregate([
    {"$group": {"_id": "$product_id", "total_quantity": {"$sum": {"$toInt": "$quantity_sold"}}}}
])

print("Popular products:")
for product in product_sales:
    print(f"Product {product['_id']}: sold {product['total_quantity']} units")
```

```python
from datetime import datetime

current_year = datetime.now().year

client_ages = []
for client in clients_collection.find():
    age = current_year - int(client['birth'])
    client_ages.append(age)

age_groups = {"< 20": 0, "20-30": 0, "30-40": 0, "40-50": 0, "50+": 0}
for age in client_ages:
    if age < 20:
        age_groups["<20"] += 1
    elif age <= 30:
        age_groups["20-30"] += 1
    elif age <= 40:
        age_groups["30-40"] += 1
    elif age <= 50:
        age_groups["40-50"] += 1
    else:
        age_groups["50+"] += 1

print("Client distribution by age group:")
for group, count in age_groups.items():
    print(f"{group} years: {count} clients")
```
